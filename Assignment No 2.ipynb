```python
try:
    number = int(input('Enter number:'))#any number upto 19 digits in it
    start = 0 
    final_number_in_words = []

    def getNotation(size):
        array = ['','','hundred','thousand','thousand','hundred thousand','million','million','hundren million',
                 'billion','billion','hundred billion','trillion','trillion','hundred trillioin','quadrillion','quadrillion',
                 'hundred quadrillion','quintillion','quintillion']
        return array[size]

    def handleOnce(no):
        num_dict = {0:'',1:'one',2:'two',3:'three',4:'four',5:'five',6:'six',7:'seven',8:'eight',9:'nine'}
        return num_dict[no]

    def handleSizeChart(length):
        size_chart = {1:1,2:2,3:3,4:1,5:2,6:3,7:1,8:2,9:3,10:1,11:2,12:3,13:1,14:2,15:3,16:1,17:2,18:3,19:1,20:2}
        return size_chart[length]

    def handleTens(no):
        if no[0]=='0':
            first = ''
            if no[1]=='0':
                second = ''
            else:
                second = handleOnce(int(no[1]))
            return first+second
        num_dict = {10:'ten',11:'eleven',12:"tweal",13:"thirteen",14:'fourteen',15:'fifteen',16:'sixteen',17:'seventeen',18:'eighteen',
                    19:'nineteen',20:'tweenty',21:'tweenty one',22:'tweenty two',23:'tweenty three',24:'tweenty four',25:'tweenty five',26:'tweenty six',
                    27:'tweenty seven',28:'tweenty eight',29:'tweenty nine',30:'thirty',31:'thirty one',32:'thirty two',
                    33:'thirty three',34:'thirty four',35:'thirty five',36:'thirty six',37:'thirty seven',38:'thirty eight',
                    39:'thirty nine',40:'fourty',41:'fourty one',42:'fourty two',43:'fourty three',44:'fourty four',45:'fourty five',
                    46:'fourty six',47:'fourty seven',48:'fourty eight',49:'fourty nine',50:'fifty',51:'fifty one',52:'fifty two',
                    53:'fifty three',54:'fifty four',55:"fifty five",56:'fifty six',57:'fifty seven',58:'fifty eight',59:'fifty nine',
                    60:'sixty',61:'sixty one',62:'sixty two',63:'sixty three',64:'sixty four',65:'sixty five',66:'sixty six',67:'sixty seven',
                    68:'sixty eight',69:'sixty nine',70:'seventy',71:'seventy one',72:"seventy two",73:'seventy three',74:'seventy four',
                    75:'seventy five',76:'seventy six',77:'seventy seven',78:'seventy eight',79:'seventy nine',80:'eighty',81:'eighty one',
                    82:'eighty two',83:'eighty three',84:'eighty four',85:'eighty five',86:'eighty six',87:'eighty seven',88:'eighty eight',
                    89:'eighty nine',90:'ninty',91:'ninty one',92:'ninty two',93:'ninty three',94:'ninty four',95:'ninty five',
                    96:'ninty six',97:'ninty seven',98:'ninty eight',99:'ninty nine'}
        return num_dict[int(no)]

    while start<len(str(number)):
        #print('value of start:',start)
        number_in_word = ''
        #print('length at top:',len(str(number)[start:])-1)
        notation = getNotation(len(str(number)[start:])-1)
        #print('length:',len(str(number)[start:]))
        size = handleSizeChart(len(str(number)[start:]))
        #print('size:',size)
        sliced_number = str(number)[start:start+size]
        #print("sliced number:",sliced_number)

        if size==1:
            #print('number')
            number_in_word = handleOnce(int(str(number)[start:start+size]))
            #print('Number_in_word:',number_in_word)
            #print('notation:',notation)
            final_number_in_words.append(number_in_word)
            final_number_in_words.append(notation)
        elif size==2:
            number_in_word = handleTens(str(number)[start:start+size])
            final_number_in_words.append(number_in_word)
            #print(notation)
            final_number_in_words.append(notation)
        elif size==3:
            first = handleOnce(int(sliced_number[0]))

            #print('first number:',int(sliced_number[0]))
            if int(sliced_number[0])!=0:
                first_notation = getNotation(len(sliced_number)-1)
                #first_notation = getNotation(len(str(number)[start:])-1)
            else:
                first_notation = ''

            second = handleTens((sliced_number[1:]))
            #print('length:::',len(str(number)[start+1:])-1)
            if int(sliced_number[1:])!=0:
                second_notation = getNotation(len(str(number)[start+1:])-1)
            else:
                second_notation = ''

            #print('second notation:',second_notation)
            #print('first:',first)
            #print('second:',second)
            number_in_words = first+" "+first_notation+" "+second+" "+second_notation
            final_number_in_words.append(number_in_words)

        start+=size  #increse the start counter
    new_final = []
    for x in final_number_in_words:
        if len(x.strip())>0:
            new_final.append(x.strip())
    print(new_final)
    print(" ".join(new_final))
except Exception as e:
    print(e)
```
