```python
book_dictionary = dict()
def insertBook(bname):
    if bname in book_dictionary.keys():
        print('\nBook already present\n')
        return False
    else:
        book_dictionary.update({bname:1})
        return True

def printBook():
    [print("\n{0}".format(key)) for key in book_dictionary.keys()]
    
def searchBook(bname):
    if bname in book_dictionary.keys():
        return True
    return False

def deleteBook(bname):
    if bname in book_dictionary.keys():
        book_dictionary.pop(bname)
        return True
    else:
        print("\nBook not present to delete\n")
        return False
        
if __name__ == '__main__':
    while True:
        print('1.Insert Book')
        print('2.Search Book')
        print('3.Print Books')
        print('4.Delete Book')
        print('0.Exit')
        operation = int(input('Enter operation:'))
        try:
            if operation == 1:
                bookname = input('Enter book name:')
                result = insertBook(bookname)
                if result:
                    print('\nBook Added Sucessfully\n')
            elif operation == 2:
                bookname = input('Enter bookname to search:')
                result = searchBook(bookname)
                if result:
                    print('\n Book is available\n')
                else:
                    print('\nBook is not available\n')
            elif operation == 3:
                printBook()
            elif operation == 4:
                bookname = input('Enter bookname to delete:')
                result = deleteBook(bookname)
                if result:
                    print('\nBook is available\n')
            elif operation not in [1,2,3,4,0]:
                print('\nInvalid operation\n')
            elif operation == 0:
                break
        except ValueError as e:
            print('\n\nPlease enter correct input\n\n')
```
